```python
import pyttsx4
```

```python
engine = pyttsx4.init()
engine.say('this is an english text to voice test.')
engine.runAndWait()
```

```python
engine.save_to_file('i am Hello World, i am a programmer. i think life is short.', 'test1.wav')
engine.runAndWait()
```
